```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib

# ファイルの読み込み
# ユーザー提供のファイル名とシート名に基づいています
file_path = '2025_診療報酬_全て統合.xlsx'
df = pd.read_excel(file_path, sheet_name='シート1')

# グラフのサイズ設定
plt.figure(figsize=(10, 6))

# 列名の設定（1列目は日付/月として扱います）
date_col = df.columns[0]
first_visit_total = '初診合計'
follow_up_total = '再診合計'

# 初診合計と再診合計の推移をプロット
plt.plot(df[date_col], df[first_visit_total], marker='o', label='初診合計')
plt.plot(df[date_col], df[follow_up_total], marker='s', label='再診合計')

# グラフの装飾
plt.title('月別 診療報酬合計推移（2025年）')
plt.xlabel('年月')
plt.ylabel('合計（点数）')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# レイアウトの調整と表示
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib

# 診療報酬ファイルを読み込む
file_name = '2025_診療報酬_全て統合.xlsx'
df = pd.read_excel(file_name, sheet_name='シート1')

# 描画サイズの設定
plt.figure(figsize=(10, 8))

# 比較したい項目のリスト（項目名と、対応する平均の列名をペアにします）
# データの項目に基づき、主要な5項目を抽出
comparison_items = [
   {'data': '初診件数'},
   {'data': '初診料'},
    {'data': '初診時間外加算'},
    {'data': '初診休日加算'},
    {'data': '初診合計'},
    {'data': '初診平均'},
    {'data': '再診件数'},
   {'data': '再診料'},
    {'data': '再診時間外加算'},
    {'data': '再診休日加算'}
    {'data': '再診合計'},
    {'data': '再診平均'},
]

# 横軸に使用する列（年月）
x_axis = df.columns[0]

# 各項目とその平均をプロット
for item, avg in comparison_items:
    # 診療項目の数値（実線・o印）
    p = plt.plot(df[x_axis], df[item], marker='o', label=item)
    color = p[0].get_color() # 現在のプロット色を取得

# グラフの設定
plt.title('診療報酬 各項目の推移')
plt.xlabel('年月')
plt.ylabel('点数')
plt.ylim(0, 5000) # 縦軸を0〜5000に設定
plt.grid(True, linestyle=':', alpha=0.7)

# 凡例をグラフの下に配置（ncol=5で横並びに調整）
# bbox_to_anchorの調整で凡例がグラフと重ならないようにしています
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.12), ncol=5, fontsize=8)

# レイアウトの調整（凡例が切れないように余裕を持たせる）
plt.tight_layout()

# PNG形式で保存
plt.savefig('medical_fee_graph.png', dpi=300, bbox_inches='tight')

# 表示
plt.show()
```
